```python
from smolagents import CodeAgent,tool,InferenceClientModel
@tool
def suggest_menu(occasion:str)->str:
    """
    Suggest menu based on occasion
    Args:
        occasion(str): The type of occasion for the party. Allowed values are:
            -"casual": For a casual get-together with friends and family.
            -"formal": For a formal dinner or event.
            -"superhero": For a superhero-themed party, inspired by the Wayne's mansion theme.
            -"custom": For a custom menu based on specific preferences or dietary restrictions.
    """
    if occasion=="casual":
        return "Pizza, snacks, and drinks."
    elif occasion=="formal":
        return "3-course dinner with wine and dessert."
    elif occasion=="superhero":
        return "Buffet with high-energy and healthy food."
    else:
        return "custom menu for butler"

agent = CodeAgent(
    tools=[suggest_menu],
    model=InferenceClientModel()
)

agent.run("prepare a custom menu for the party.")
```

```python
from smolagents import CodeAgent, DuckDuckGoSearchTool, InferenceClientModel, FinalAnswerTool, tool,Tool, VisitWebpageTool
@tool
def suggest_menu(occasion: str) -> str:
    """
    Suggests a menu based on the occasion.
    Args:
        occasion: The type of occasion for the party.
    """
    if occasion == "casual":
        return "Pizza, snacks, and drinks."
    elif occasion == "formal":
        return "3-course dinner with wine and dessert."
    elif occasion == "superhero":
        return "Buffet with high-energy and healthy food."
    else:
        return "Custom menu for the butler."

@tool
def catering_service_tool(query: str) -> str:
    """
    This tool returns the highest-rated catering service in Gotham City.
    
    Args:
        query: A search term for finding catering services.
    """
    # Example list of catering services and their ratings
    services = {
        "Gotham Catering Co.": 4.9,
        "Wayne Manor Catering": 4.8,
        "Gotham City Events": 4.7,
    }
    
    # Find the highest rated catering service (simulating search query filtering)
    best_service = max(services, key=services.get)
    
    return best_service


class SuperheroPartyThemeTool(Tool):
    name = "superhero_party_theme_generator"
    description = """
    This tool suggests creative superhero-themed party ideas based on a category.
    It returns a unique party theme idea."""
    
    inputs = {
        "category": {
            "type": "string",
            "description": "The type of superhero party (e.g., 'classic heroes', 'villain masquerade', 'futuristic Gotham').",
        }
    }
    
    output_type = "string"

    def forward(self, category: str):
        themes = {
            "classic heroes": "Justice League Gala: Guests come dressed as their favorite DC heroes with themed cocktails like 'The Kryptonite Punch'.",
            "villain masquerade": "Gotham Rogues' Ball: A mysterious masquerade where guests dress as classic Batman villains.",
            "futuristic Gotham": "Neo-Gotham Night: A cyberpunk-style party inspired by Batman Beyond, with neon decorations and futuristic gadgets."
        }
        
        return themes.get(category.lower(), "Themed party idea not found. Try 'classic heroes', 'villain masquerade', or 'futuristic Gotham'.")


# Creating the agent with the new tools
agent=CodeAgent(
    tools=[DuckDuckGoSearchTool(),
           VisitWebpageTool(),
           suggest_menu,
           catering_service_tool,
           SuperheroPartyThemeTool(),
           FinalAnswerTool()],
    model=InferenceClientModel(),
    max_steps=10,
    verbosity_level=2
)

agent.run("Give me the best playlist for a party at the Wayne's mansion. The party idea is a 'villain masquerade' theme")
```

```python
# stock performance tool
from smolagents import tool,Tool,DuckDuckGoSearchTool,VisitWebpageTool, FinalAnswerTool, InferenceClientModel, CodeAgent
import yfinance as yf
import matplotlib.pyplot as plt
@tool
def stock_performance_tool(ticker: str,time_period: str) -> str:
    """
    Returns the performance data for a given stock ticker.
    
    Args:
        ticker: The stock ticker symbol.
        time_period: The time period for which to display performance data. input should be in the format accepted by yfinance (e.g., '1d', '5d', '1mo', '3mo', '6mo', '1y', etc.)
    """
    stock = yf.Ticker(ticker)
    hist = stock.history(period=time_period)
    
    if hist.empty:
        return f"No data available for {ticker}"
    
    plt.figure(figsize=(10, 5))
    plt.plot(hist.index, hist['Close'])
    plt.title(f"{ticker} Stock Price")
    plt.xlabel("Date")
    plt.ylabel("Price")
    plt.show()
    
    return f"Performance data for {ticker} displayed in the plot."
@tool
def stock_cat_tool(ticker: str) -> str:
    """
    Returns a brief category and description of the stock.
    
    Args:
        ticker: The stock ticker symbol.
    """
    stock = yf.Ticker(ticker)
    info = stock.info
    
    category = info.get('sector', 'Unknown Sector')
    description = info.get('longBusinessSummary', 'No description available.')
    
    return f"{ticker} belongs to the {category} sector. Description: {description}"

agent=CodeAgent(
    tools=[stock_performance_tool, stock_cat_tool, DuckDuckGoSearchTool(), VisitWebpageTool(), FinalAnswerTool()],
    model=InferenceClientModel(),
    max_steps=10,
    verbosity_level=2
)

agent.run("What is the stock performance of AAPL in the last 6 months? Also, give me a brief description of the company.")
```

```python
from ollama import Client
from smolagents import CodeAgent
from smolagents.models import ChatMessage

class OllamaModel:
    def __init__(self, model_name="mistral"):
        self.model_name = model_name
        self.client = Client(host="http://localhost:11434")

    def __call__(self, messages, **kwargs):
        ollama_messages = []
        for msg in messages:
            if hasattr(msg, "role"):
                content = msg.content if isinstance(msg.content, str) else str(msg.content)
                ollama_messages.append({"role": msg.role, "content": content})
            else:
                ollama_messages.append(msg)

        response = self.client.chat(
            model=self.model_name,
            messages=ollama_messages
        )
        content = response["message"]["content"]
        return ChatMessage(role="assistant", content=content)

    def generate(self, messages, **kwargs):
        return self.__call__(messages, **kwargs)

agent = CodeAgent(
    tools=[],
    model=OllamaModel(model_name="mistral")
)
agent.run("What is the best music for a party at Wayne's mansion? The theme is 'villain masquerade'.")
```
